```python
#important packages vectorizing
import pandas as pd
import numpy as np
import nltk
from ast import literal_eval
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize
```

```python
#importing titles
df = pd.read_csv('datasets/preprocessed-data2.csv')
df.head()
```

```python
suitable_topic = df.loc[:,['Title']] 
suitable_topic.iloc[2]['Title']
```

```python
def apply_clustering(document):
    document  = literal_eval(document)
    vectorizer = TfidfVectorizer(stop_words='english')
        
    tfidVec = vectorizer.fit_transform(document)
    
    #finding suitable topics
    suitable_topics = suitable_topic
    suitable_topics = suitable_topics.values.reshape(-1,).tolist()
    suitable_topics = [x for x in suitable_topics if str(x) != 'nan']
        
    #applying k-means and performing clustering
    kmeans = KMeans(n_clusters=12, init='k-means++', max_iter=500, n_init=1)
    fitted = kmeans.fit(tfidVec)
    
    something= suitable_topics
    prediction = kmeans.predict(vectorizer.transform(something))
    
    max_index_col = np.argmax(prediction, axis=0)
    if prediction[max_index_col] == 0:
        return "Sorry result not found"
    else:
        return suitable_topic.iloc[max_index_col]['Title']

    
```

```python
# %timeit result = df.loc[:,'Abstract'].apply(lambda x: apply_clustering(x))
result = df.loc[:,'PaperText'].apply(lambda x: apply_clustering(x))
```

```python
df['result'] = result
df.head(20)
```
